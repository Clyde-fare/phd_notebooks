```python
import numpy as np

def encode(rxyz_data):
    """Encodes a string of xyz data to vector form for ML algorithms"""

    xyz_data = [l.strip().split() for l in rxyz_data[2:]]
    xyz_coords = np.array([l[1:] for l in xyz_data], dtype=np.float32)

    c_array = np.array([int(l[0]=='C') for l in xyz_data])
    h_array = np.array([int(l[0]=='H') for l in xyz_data])
    n_array = np.array([int(l[0]=='N') for l in xyz_data])
    
    valid_atoms = (c_array+h_array+n_array).astype(np.bool)
    
    if not all(valid_atoms):
        raise RuntimeError('Unidentified atom in {a}'.format(a=np.array(xyz_data)[np.invert(valid_atoms)]))
        
    r_encoded_data = np.hstack((c_array[:,np.newaxis], 
                                h_array[:,np.newaxis], 
                                n_array[:,np.newaxis],
                                xyz_coords))
        
    return r_encoded_data.ravel()

def decode(encoded_data):
    """Decodes a vector representing a structure to an xyz string"""

    no_atoms = encoded_data.shape[1]/6
    encoded_data = encoded_data.reshape([no_atoms,6])
    xyz_data=[]
    
    for a in encoded_data:
        if a[0]==1:
            c='C'
        elif a[1]==1:
            c='H'
        elif a[2]==1:
            c='N'
        else:
            raise RuntimeError("This can't happen")

        atom_xyz = [c] + list(map(str, a[3:]))
        xyz_data.append(" ".join(atom_xyz))
        
    return "{no_c}\n\n".format(no_c=no_atoms) + "\n".join(xyz_data)
```

```python
from sklearn.tree import DecisionTreeRegressor

structure_fns = ['p{n}.xyz'.format(n=i) for i in range(8)]
xyz_data = []

for fn in structure_fns:
    with open(fn) as f:
        xyz_data.append(f.readlines())

structure_data = [encode(xyz_d) for xyz_d in xyz_data][4:-1]
no_structures = len(structure_data)

# Create a dataset
X = np.arange(no_structures)[:,np.newaxis]
y = np.array(structure_data)

# Fit regression model
clf_1 = DecisionTreeRegressor(max_depth=2)
clf_2 = DecisionTreeRegressor(max_depth=5)
clf_3 = DecisionTreeRegressor(max_depth=8)

clf_1.fit(X, y)
clf_2.fit(X, y)
clf_3.fit(X, y)
```

```python
clf_1.score(X,y)
```

```python
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt

###############################################################################
# Generate sample data
X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(X).ravel()

###############################################################################
# Add noise to targets
y[::5] += 3 * (0.5 - np.random.rand(8))

###############################################################################
# Fit regression model
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
svr_lin = SVR(kernel='linear', C=1e3)
svr_poly = SVR(kernel='poly', C=1e3, degree=2)
y_rbf = svr_rbf.fit(X, y).predict(X)
y_lin = svr_lin.fit(X, y).predict(X)
y_poly = svr_poly.fit(X, y).predict(X)

###############################################################################
# look at the results
plt.scatter(X, y, c='k', label='data')
plt.hold('on')
plt.plot(X, y_rbf, c='g', label='RBF model')
plt.plot(X, y_lin, c='r', label='Linear model')
plt.plot(X, y_poly, c='b', label='Polynomial model')
plt.xlabel('data')
plt.ylabel('target')
plt.title('Support Vector Regression')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt

###############################################################################
# Generate sample data
X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.array([np.sin(X).ravel(), np.cos(X).ravel()]).T

###############################################################################
# Add noise to targets
y[::5] += 3 * (0.5 - np.random.rand(8,2))

###############################################################################
# Fit regression model
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
svr_lin = SVR(kernel='linear', C=1e3)
svr_poly = SVR(kernel='poly', C=1e3, degree=2)
y_rbf = svr_rbf.fit(X, y).predict(X)
y_lin = svr_lin.fit(X, y).predict(X)
y_poly = svr_poly.fit(X, y).predict(X)

###############################################################################
# look at the results
plt.scatter(X, y[:,0], c='k', label='data')
plt.hold('on')
plt.plot(X, y_rbf[:,0], c='g', label='RBF model')
plt.plot(X, y_lin[:,0], c='r', label='Linear model')
plt.plot(X, y_poly[:,0], c='b', label='Polynomial model')
plt.xlabel('data')
plt.ylabel('target')
plt.title('Support Vector Regression')
plt.legend()
plt.show()
```
