Search for cis-trans transition state by scanning dihedral involving the double bond:

```python
scan_m = read(p_coumarics[0].calc.label +'.log')
scan_m.set_calculator(Gaussian(label='p_coumaric_scan', method='B3LYP', basis='6-31G*', opt='modredundant', nosymmetry='nosymmetry'))
scan_m.calc.set_redundant(coord_type ='dihedral', atom_nos = [4, 2, 1, 5], action_type='add')
scan_m.calc.set_redundant(coord_type ='dihedral', atom_nos = [10, 2, 1, 16], action_type='scan', frames=30, degrees=6)
scan_m.calc.set_job(nodes=8, memory=8*1400, time=5, queue='pqmb', version='g09')   
```

```python
scan_m2 = read(p_coumarics[0].calc.label +'.log')
scan_m2.set_calculator(Gaussian(label='p_coumaric_scan_2', method='B3LYP', basis='6-31G*', opt='modredundant', nosymmetry='nosymmetry'))
scan_m2.calc.set_redundant(coord_type ='dihedral', atom_nos = [4, 2, 1, 5], action_type='remove')
scan_m2.calc.set_redundant(coord_type ='dihedral', atom_nos = [10, 2, 1, 16], action_type='scan', frames=30, degrees=6)
scan_m2.calc.set_job(nodes=8, memory=8*1400, time=5, queue='pqmb', version='g09')
```

The problem here is that we reach a point in the scan where relaxing the other coordinates means the three of the atoms that define the dihedral we are scanning across become linear, this breaks the definition of the dihedral and consequently breaks the scan. One of the hydrogens (20) is getting stuck at the side, so we will try locking it in place by removing the relevant dihedral from the optimisation.

```python
check_calcs_v2([scan_m, scan_m2], data_file='init_scan')
mols_to_html([scan_m, scan_m2])
```

The problem occurs close to the half way point so we will construct a 360 degree scan by moving in both directions from the two end points:

```python
scans = []
for i in [1,2]:
    scan = read(p_coumarics[0].calc.label +'.log')
    scan.set_calculator(Gaussian(label='p_coumaric_scan_{n}'.format(n=2+i), method='B3LYP', basis='6-31G*', opt='modredundant', nosymmetry='nosymmetry'))
    scan.calc.set_redundant(coord_type ='dihedral', atom_nos = [20, 1, 16, 17], action_type='freeze')
    scan.calc.set_redundant(coord_type ='dihedral', atom_nos = [10, 2, 1, 16], action_type='scan', frames=30, degrees=6*-1**i)
    scan.calc.set_job(nodes=8, memory=8*1400, time=5, queue='pqmb', version='g09')

    scan_2= read(p_coumarics[1].calc.label +'.log')
    scan_2.set_calculator(Gaussian(label='p_coumaric_scan_{n}'.format(n=4+i), method='B3LYP', basis='6-31G*', opt='modredundant', nosymmetry='nosymmetry'))
    scan_2.calc.set_redundant(coord_type ='dihedral', atom_nos = [20, 1, 16, 17], action_type='freeze')
    scan_2.calc.set_redundant(coord_type ='dihedral', atom_nos = [10, 2, 1, 16], action_type='scan', frames=30, degrees=6*-1**i)
    scan_2.calc.set_job(nodes=8, memory=8*1400, time=5, queue='pqmb', version='g09')
    
    scan_3= read(p_coumarics[1].calc.label +'.log')
    scan_3.set_calculator(Gaussian(label='p_coumaric_scan_{n}'.format(n=6+i), method='B3LYP', basis='6-31G*', opt='modredundant', nosymmetry='nosymmetry'))
    scan_3.calc.set_redundant(coord_type ='dihedral', atom_nos = [3, 2, 4, 14], action_type='freeze')
    scan_3.calc.set_redundant(coord_type ='dihedral', atom_nos = [15, 5, 6, 12], action_type='freeze')
    scan_3.calc.set_redundant(coord_type ='angle', atom_nos = [2, 4, 5], action_type='freeze')
    scan_3.calc.set_redundant(coord_type ='angle', atom_nos = [4, 5, 6], action_type='freeze')
    scan_3.calc.set_redundant(coord_type ='dihedral', atom_nos = [6, 5, 4, 2], action_type='scan', frames=30, degrees=6*-1**i)
    scan_3.calc.set_job(nodes=8, memory=8*1400, time=5, queue='pqmb', version='g09')
    
    scans.append(scan)
    scans.append(scan_2)
    scans.append(scan_3)
    
scan_4= read(p_coumarics[1].calc.label +'.log')
scan_4.set_calculator(Gaussian(label='p_coumaric_scan_{n}'.format(n=9), method='B3LYP', basis='6-31G*', opt='modredundant', nosymmetry='nosymmetry'))
scan_4.calc.set_redundant(coord_type ='dihedral', atom_nos = [3, 2, 4, 14], action_type='freeze')
scan_4.calc.set_redundant(coord_type ='dihedral', atom_nos = [15, 5, 6, 12], action_type='freeze')
scan_4.calc.set_redundant(coord_type ='angle', atom_nos = [2, 4, 5], action_type='freeze')
scan_4.calc.set_redundant(coord_type ='angle', atom_nos = [4, 5, 6], action_type='freeze')
scan_4.calc.set_redundant(coord_type ='dihedral', atom_nos = [15, 5, 4, 14], action_type='scan', frames=30, degrees=6)
scan_4.calc.set_job(nodes=8, memory=8*1400, time=5, queue='pqmb', version='g09')
scans.append(scan_4)
```

```python
cdict = {'method':['B3LYP','CIS','CIS'], 
         'irc':['(gradientonly, downhill, maxpoints=100)', '(calcfc, downhill, recorrect=test, maxpoints=50)', '(gradientonly, downhill, maxpoints=100)'], 
         'label_id':['_1', '_1_CIS', '_1b_CIS']}

photos=[]
for i in range(3):
    photo = read(p_coumarics[0].calc.label +'.log')
    photo.set_calculator(Gaussian(label='p_coumaric_photo' + cdict['label_id'][i], method=cdict['method'][i], basis='6-31G*', irc=cdict['irc'][i], td='root=1',nosymmetry='nosymmetry'))
    photo.calc.set_job(nodes=8, memory=8*1400, time=5, queue='pqmb', version='g09')
    photos.append(photo)  

cdict['label_id']= ['_2', '_2_CIS', '_2b_CIS']
for i in range(3):
    photo = read(p_coumarics[1].calc.label +'.log')
    photo.set_calculator(Gaussian(label='p_coumaric_photo' + cdict['label_id'][i], method=cdict['method'][i], basis='6-31G*', irc=cdict['irc'][i], td='(root=1)',nosymmetry='nosymmetry'))
    photo.calc.set_job(nodes=16, memory=16*1400, time=5, queue='pqmb', version='g09')
    photos.append(photo) 
```

Single negative frequency for CIS isomer corresponds to twisting about the double bond, displace along this coordinate to get us off the saddle point then restart the IRC, TRANS isomer has no negative frequency (weird) but 1st norm mode also corresponds to twisting about the double bond so we will again displace along this cooord and restart the IRC.

```python
cis_pos_delta, trans_pos_delta = array(freqs[0].calc.max_data['vibdisps'][0]), array(freqs[1].calc.max_data['vibdisps'][0])
```

```python
new_cis_photo = read(freqs[0].calc.label +'.log')
new_cis_photo.set_positions(new_cis_photo.get_positions() + cis_pos_delta)
new_cis_photo.set_calculator(Gaussian(label='p_coumaric_photo_1_CIS_cont', method='CIS', basis='6-31G*', irc='(calcfc, downhill, recorrect=test, maxpoints=50)', td='(root=1)',nosymmetry='nosymmetry', guess='read', oldchk=freqs[0].calc.label + '.chk'))
new_cis_photo.calc.set_job(nodes=16, memory=16*1400, time=10, queue='pqmb', version='g09')

new_trans_photo = read(freqs[1].calc.label +'.log')
new_trans_photo.set_positions(new_trans_photo.get_positions() + trans_pos_delta)
new_trans_photo.set_calculator(Gaussian(label='p_coumaric_photo_1_TRANS_cont', method='CIS', basis='6-31G*', irc='(calcfc, downhill, recorrect=test, maxpoints=50)', td='(root=1)',nosymmetry='nosymmetry', guess='read', oldchk=freqs[0].calc.label + '.chk'))
new_trans_photo.calc.set_job(nodes=16, memory=16*1400, time=10, queue='pqmb', version='g09')
```
